```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
df['genres_clean']=df['genres'].str.replace('|',' ',regex=False)
df['genres_clean']
```

```python
genres_tfidf=tfidf.fit_transform(df['genres'])
```

```python
genres_tfidf
```

```python
genre_df=pd.DataFrame(
    genres_tfidf.toarray(),
    columns=tfidf.get_feature_names_out()
)
```

```python
df1=pd.concat([df,genre_df],axis=1)
```

```python
df1
```

```python
df1=df1.drop(columns='genres')
```

```python
df1
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix=cosine_similarity(genres_tfidf)
```

```python
similarity_matrix.shape
```

```python
indices=pd.Series(df1.index,index=df1['title']).drop_duplicates()
```

```python
from operator import index
def recommend(title,top_n=5):
  if title not in indices:
    return "Movie Not Found"

  idx=indices[title]

  # Get similarity scores
  sim_scores=list(enumerate(similarity_matrix[idx]))

  # Sort movies based on similarity scores
  sim_scores=sorted(sim_scores,key=lambda x:x[1],reverse=True)

  # Skip itself
  sim_scores=sim_scores[1:top_n+1]

  movie_indices=[i[0] for i in sim_scores]

  return df1[['title','genres_clean']].iloc[movie_indices]
```

```python
recommend("Toy Story (1995)")
```
